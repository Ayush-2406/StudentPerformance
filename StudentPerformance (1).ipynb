```
#importing packages.
#importing the datasets.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data_load=pd.read_csv("http://bit.ly/w-data")
print("Successsfully Loaded")


```

```
#We can see the first six rows of dataset.
data_load.head(6)

```

# Data Visualization

```
# We are plotting the graph on the basis of datasets.
data_load.plot(x='Hours',y='Scores',style='o')
plt.title("HOURS VS PERCENTAGE")
plt.xlabel("HOURS STUDIED")
plt.ylabel("PERCENTAGE SCORED")
plt.show()
```

The above graph shows that there is a linear regression.

# Data Preparation

```
X=data_load.iloc[:,:-1].values
Y=data_load.iloc[:,1].values

X.reshape(-1,1)
Y.reshape(-1,1)
```

# Splitting DataSet

```
#importing the packeage to split the datasets.
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
```

# Training the Model

```
#import the supervised LinearRegression Model and the split data has been trained.
from sklearn.linear_model import LinearRegression
r=LinearRegression()
r.fit(X_train,Y_train)
```

```
#Plotting the garph of trained model
plt.scatter(X_train,Y_train)
plt.title("Training")
plt.plot(X_train,r.predict(X_train))
plt.xlabel('HOURS STUDIED')
plt.ylabel('PERCENTAGE SCORED')
plt.show()
```

```
#Accuracy
r.score(X_train,Y_train)
```

```
#Deriving the Regression line equation
line=r.coef_*X+r.intercept_
plt.scatter(X,Y)
plt.title("Regression Line")
plt.plot(X,line)
plt.xlabel('HOURS STUDIED')
plt.ylabel('PERCENTAGE SCORED')
plt.show()
```

## PREDICTION

```
#Printig the trained value and Predicting the Value from the Dataset. 
print(X_test)
Y_pred=r.predict(X_test)
```

```
# Showing the actual value and Predicted Value.
df=pd.DataFrame({'Actual':Y_test,'Predict':Y_pred})
df
```

```
#Predicted value on the basis of trained Model.
Hours=[[9.25]]
own_pred=r.predict(Hours)
print("Number of Hours:{}".format(Hours))
print("Prediction of Score:{}".format(own_pred[0]))
```

# Evaluation

```
#Accuracy
r.score(X_test,Y_test)
```

```
#import metrics from sklearn

from sklearn import metrics
print("Mean Absolute Error:",metrics.mean_absolute_error(Y_test,Y_pred))
```
